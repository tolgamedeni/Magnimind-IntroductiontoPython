```python
#Question09
#To merge I am using join
from functools import reduce
flights = pd.read_csv('flights.csv')

airline = pd.read_csv('airlines.csv')

dfs = [flights,airline]
#carrier is the common on both dataframes, like making an outer join in the databases table, we applied it in
#merge all DataFrames into one
final_df =reduce(lambda  left,right: pd.merge(left,right,on=['carrier'], how='outer'), dfs).fillna('none')

#view merged DataFrame
final_df

```

```python
#Question 10, to be able checked, I used the following methods, it seems, there is not any missing values
final_df.isnull().any()
final_df.notnull()
```

```python
#Question 11
import pandas as pd
from functools import reduce
state_populations = pd.read_csv('state-populations.csv')

census_divisions = pd.read_csv('census-divisions.csv')

dfs = [state_populations,census_divisions]
#carrier is the common on both dataframes, like making an outer join in the databases table, we applied it in
#merge all DataFrames into one
census =reduce(lambda  left,right: pd.merge(left,right,on=['state'], how='outer'), dfs).fillna('none')

#view merged DataFrame
census
#Question12
census_year=census.iloc[:,1:8].sum()
census_year=census_year.to_frame()
index=census_year.index
index.name="year"
census_year.columns=["population"]
census_year
```
