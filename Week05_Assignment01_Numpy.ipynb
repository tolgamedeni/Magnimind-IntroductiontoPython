```python
#Week05_Numpy Assignment
#Question01-Question02
import numpy as np
house_size=[50,75,90,120,100,90,60,50,120,100]
house_room=[1,2,3,4,3,3,1,1,4,3]
house_price=[50000,75000,90000,120000,100000,90000,60000,50000,120000,100000]
#len(house_size)
#creating an empty array
house_combine=[[0 for j in range(3)]for i in range(10)]
counter=0
for i in range(10):
        house_combine[i][0]=house_size[i]
        house_combine[i][1]=house_room[i]
        house_combine[i][2]=house_price[i]
#This will already created the desired transposed array
house_combine_array=np.asarray(house_combine)
print("Created array with including three lists:\n",house_combine_array)
house_combine_array=house_combine_array.transpose()
print("Transposed array:\n",house_combine_array)
house_combine_array=house_combine_array.transpose()
print("Re-Transposed array:\n",house_combine_array)
```
